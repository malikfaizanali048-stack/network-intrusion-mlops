```python
import pandas as pd
import numpy as np

def load_and_clean(filepath):
    df = pd.read_csv(filepath, encoding='cp1252')
    df.columns = df.columns.str.strip()
    df.replace([np.inf, -np.inf], np.nan, inplace=True)
    before = df.shape[0]
    df.dropna(inplace=True)
    after = df.shape[0]
    print(f"{filepath}")
    print(f"  Rows before: {before}, after: {after}, dropped: {before - after}")
    print(f"  Label distribution:\n{df['Label'].value_counts()}\n")
    return df
```

```python
tuesday = load_and_clean('../data/raw/Tuesday-WorkingHours.pcap_ISCX.csv')
```

```python
wednesday = load_and_clean('../data/raw/Wednesday-workingHours.pcap_ISCX.csv')
```

```python
thursday_morning = load_and_clean('../data/raw/Thursday-WorkingHours-Morning-WebAttacks.pcap_ISCX.csv')
thursday_morning['Label'] = thursday_morning['Label'].str.replace('ï¿½', '-', regex=False)
print(thursday_morning['Label'].value_counts())
```

```python
thursday_afternoon = load_and_clean('../data/raw/Thursday-WorkingHours-Afternoon-Infilteration.pcap_ISCX.csv')
```

```python
friday_morning = load_and_clean('../data/raw/Friday-WorkingHours-Morning.pcap_ISCX.csv')
```

```python
friday_afternoon_portscan = load_and_clean('../data/raw/Friday-WorkingHours-Afternoon-PortScan.pcap_ISCX.csv')
```

```python
friday_afternoon_ddos = load_and_clean('../data/raw/Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv')
```

```python
all_data = pd.concat([
    tuesday, wednesday, thursday_morning, thursday_afternoon,
    friday_morning, friday_afternoon_portscan, friday_afternoon_ddos
], ignore_index=True)

print("Combined shape:", all_data.shape)
print("\nFull label distribution:\n", all_data['Label'].value_counts())
```

```python
all_data['Label_Binary'] = all_data['Label'].apply(lambda x: 0 if x == 'BENIGN' else 1)
print("Binary label distribution:")
print(all_data['Label_Binary'].value_counts())
```

```python
label_counts = all_data['Label'].value_counts()
rare_labels = label_counts[label_counts < 100].index.tolist()
print("Rare labels being grouped into 'Other':", rare_labels)

all_data['Label_Multiclass'] = all_data['Label'].apply(
    lambda x: 'Other' if x in rare_labels else x
)
print("\nMulticlass label distribution:")
print(all_data['Label_Multiclass'].value_counts())
```

```python
print("Duplicate rows:", all_data.duplicated().sum())
print("Shape before removing duplicates:", all_data.shape)

all_data.drop_duplicates(inplace=True)

print("Shape after removing duplicates:", all_data.shape)
print("\nBinary label distribution after dedup:")
print(all_data['Label_Binary'].value_counts())
```

```python
all_data.to_csv('../data/processed_combined_final.csv', index=False)
print("Saved final processed dataset")
```
